```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salarios_marcos = np.array([100,200,400,300])
salarios_gileno = np.array([50,300,500,450])
```

```python
#adicionando label
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label='Salário Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='s', ms=8, label='Salário Gileno')
```

```python
#colocando legenda
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label='Salário Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='s', ms=8, label='Salário Gileno')
plt.legend()
plt.show()
```

```python
#colocando legenda
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label='Salário Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8, label='Salário Gileno')
plt.legend()
plt.show()
```

```python
#colocando legenda
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label='Salário Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8, label='Salário Gileno')
plt.legend(loc='upper left')
plt.show()
```

```python
#colocando mais uma linha
salarios_pedro = [200,150,300,700]
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label='Salário Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8, label='Salário Gileno')
plt.plot(salarios_pedro, c='Green', ls='--', marker='o', ms=8, label='Salário Pedro')
plt.legend(loc='upper left')
plt.show()
```
